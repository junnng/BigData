```python
import matplotlib.pyplot as plt

plt.title("Players Goal and Assists")

goals=[13,9,8,4,3]
assists=[2,9,4,2,0]
players=['kane','son','Alli','Moura','Bergwin']
plt.scatter(goals,assists)

for goal,assist,player in zip(goals,assists,players):
    plt.annotate(player,xy=(goal,assist),xytext=(5,-5),textcoords="offset points")
    
plt.xlim(0,15)
plt.xlabel('Goals')
plt.ylabel("Assists")
plt.show()
```

```python
import csv

f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f,delimiter=',')

next(data)

max_temp = -999
max_date=''

for row in data:
    if row[4]=='':
        row[4]=-999
    row[4]=float(row[4])
    
    if max_temp < row[4]:
        max_temp = row[4]
        max_date = row[0]

f.close()
print('기상 관측 아래 서울의 최고 기온 : ',max_temp,'도 입니다.')
print('기상 관측 아래 서울의 최고 기온 날짜는 :',max_date,'입니다.')
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f,delimiter=',')

next(data)

result=[]
low=[]

for row in data:
    if row[4] !='' and row[3] !='':
        if 1999 <= int(row[0].split('-')[0]):
          if row[0].split('-')[1]=='11' and row[0].split('-')[2]=='08':
            result.append(float(row[4]))
            low.append(float(row[3]))
        
plt.rc('font',family='Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')
plt.plot(result,'pink')
plt.plot(low,'lightblue')
plt.legend()
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv')
data=csv.reader(f)

man=[]
woman=[]

for row in data:
    if '대연제3동' in row[0]:
        for i in row[3:104]:
            man.append(-int(i))
        for i in row[106:]:
            woman.append(int(i))
f.close()

plt.rc('font',family="Malgun Gothic")
plt.title('대연제3동 지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus']=False
plt.barh(range(101),man,color='b',label="남성")
plt.barh(range(101),woman,color='r',label="여성")
plt.legend()
plt.show()

        
```

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv')
data=csv.reader(f)

man=[]
woman=[]

name= input('찾고 싶은 지역의 이름을 입력하세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            man.append(-int(i))
        for i in row[106:]:
            woman.append(int(i))
f.close()

plt.rc('font',family="Malgun Gothic")
plt.title(name,'지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus']=False
plt.barh(range(101),man,color='b',label="남성")
plt.barh(range(101),woman,color='r',label="여성")
plt.legend()
plt.show()

        
```
