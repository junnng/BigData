```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

result=[]

for row in data:
    row[:4]=map(int,row[4:])
    result.append(sum(row[10:15:2]))
    
import matplotlib.pyplot as plt
result.sort()
plt.bar(range(len(result)),result)
plt.show()
```

```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

mx=0
mx_station=''

t=int(input('입력한 시간대의 승차 인원이 가장 많은 역? '))

for row in data:
    row[4:]=map(int,row[4:])
    a=row[4+(t-4)+2]
    if a >mx:
        mx=a;
        mx_station=row[3]+'('+row[1]+')'
        
print(mx_station,mx)
```

```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

mx=[0] *24
mx_station=['']*24
temp_str=''

print(mx)
print(mx_station)

for row in data:
    row[4:]=map(int,row[4:])
    for t in range(24):
        a = row[t*2+4]
        if a >mx[t]:
            mx[t]=a;
            if t+4 >24:
                tempt_str='('+str(t+4-24)+')'
            else:
                  tempt_str='('+str(t+4)+')'    
            mx_station[t]=row[3]+temp_str
print(mx_station,mx)
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24),mx)
plt.xticks(range(24),mx_station,rotation=90)
plt.show()
```

```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

s_in=[0]*24
s_out=[0]*24

for row in data:
    row[4:]=map(int,row[4:])
    for i in range(24):
        s_in[i]=s_in[i]+row[4+i*2]
        s_out[i]=s_out[i]+row[5+i*2]
        
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in,label='승차')
plt.plot(s_out,label='하차')
plt.legend()
plt.xticks(range(24),range(4,28))
plt.show()
```
