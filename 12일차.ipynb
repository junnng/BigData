```python
t_df=pd.read_csv("train.csv")

alive=t_df[t_df["Survived"]==1]
dead=t_df[t_df['Survived']==0]

print(len(alive),'/',len(t_df))
print(len(dead),'/',len(t_df))

plt.bar(['alive','dead'],height=[len(alive),len(dead)])
plt.show()
```

```python
t_df=pd.read_csv("train.csv")

alive=t_df[t_df["Survived"]==1]
dead=t_df[t_df['Survived']==0]

plt.scatter(alive["PassengerId"],alive["Fare"],color="blue")
plt.scatter(dead["PassengerId"],dead["Fare"],color="red")
plt.xlabel("PassengerId")
plt.ylabel("Fare")
plt.show()
```

```python
over_50=t_df[t_df["Fare"]>=50]
under_50=t_df[t_df["Fare"]<50]

print(len(over_50))
print(len(under_50))
```

```python
print(len(over_50),'/',len(t_df),str(len(over_50)/len(t_df)*100),"%")
print(len(under_50),'/',len(t_df),str(len(under_50)/len(t_df)*100),"%")
```

```python
alive_over_50=over_50[over_50["Survived"]==1]
alive_under_50=under_50[under_50['Survived']==0]

print(len(over_50),'/',len(t_df),str(len(alive_over_50)/len(over_50)*100)[:5],"%")
print(len(under_50),'/',len(t_df),str(len(alive_under_50)/len(under_50)*100)[:5],"%")
```
